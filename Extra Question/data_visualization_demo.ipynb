# Data Visualization with Pandas and Matplotlib

This notebook demonstrates how to read data from a CSV file using Pandas and create visualizations using Matplotlib.

```python

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import csv
import random

# File name for the CSV
file_name = "sample_data.csv"

# Generate 100 entries
with open(file_name, mode="w", newline="") as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(["Category", "Value"])
    # Write 100 rows of data
    for i in range(1, 101):
        category = f"Category_{i}"
        value = random.randint(1, 100)  # Random value between 1 and 100
        writer.writerow([category, value])

print(f"File '{file_name}' created with 100 entries.")
```

### Step 1: Create or Load a CSV File

```python

# Define the path to the CSV file
csv_path = "sample_data.csv"

# Load the data from the CSV file
data = pd.read_csv(csv_path)

# Display the first few rows of the data
data.head()
```

### Step 2: Bar Chart

```python

# Create a bar chart to visualize the data
plt.figure(figsize=(8, 5))
plt.bar(data["Category"], data["Value"], color='skyblue')
plt.title("Bar Chart of Values by Category")
plt.xlabel("Category")
plt.ylabel("Value")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Step 3: Pie Chart

```python

# Create a pie chart to visualize the data
plt.figure(figsize=(8, 8))
plt.pie(data["Value"], labels=data["Category"], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Pie Chart of Values by Category")
plt.tight_layout()
plt.show()
```
